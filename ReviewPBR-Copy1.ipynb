```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
import pandas as pd
import numpy as np
import nltk  
import numpy as np  
import random  
import string
import re
from itertools import repeat
from keras.models import Sequential
from keras.layers import Dense
import keras
from keras import backend as K
from keras import regularizers
```

```python
df = pd.DataFrame()
df  = pd.read_csv('Atrain_0.01.csv',encoding='utf-8',header=None)
df.head()
DR = np.asarray(df)
print("Matrix of dimension: (",DR.shape[0],",",DR.shape[1],")")
```

```python
df = pd.DataFrame()
df  = pd.read_csv('Ctest_0.01.csv',encoding='utf-8',header=None)
df.head()
DRT = np.asarray(df)
print("Matrix of dimension: (",DRT.shape[0],",",DRT.shape[1],")")
```

```python
df = pd.DataFrame()
df  = pd.read_csv('nYtrain.csv',encoding='utf-8',header=None)
Y_train = df[0]
Y_train[Y_train == -1] = 0
df.head()
```

```python
df = pd.DataFrame()
df  = pd.read_csv('nYtest.csv',encoding='utf-8',header=None)
Y_test = df[0]
Y_test[Y_test == -1] = 0
df.head()
```

```python
print(Y_train.shape)
print(Y_test.shape)
```

```python
# Create NN model
num_class = 2
model = Sequential()
# Add model layers
#model.add(Dense(10, activation='sigmoid', input_shape=(num_class,)))


model.add(Dense(10, input_dim=num_class,
                kernel_regularizer=regularizers.l2(1e-14),
                activity_regularizer=regularizers.l1(1e-14)))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
```

```python
#opt = keras.optimizers.SGD(decay=1e-14)
model.compile(loss='binary_crossentropy', optimizer='SGD', metrics=['accuracy'])
```

```python
model.fit(DR, Y_train, epochs=150, batch_size=10, validation_data=(DRT, Y_test))
#del model
K.clear_session()
```
